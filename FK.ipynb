```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the drift and diffusion functions
def mu(X,t):
    return 0.1*X 
def sigma(X,t):
    return 0.3*X
#Initial condition
X0 = 1.0
#Discretize time
T = 1.0
N = 1000
dt = T/N
t = np.linspace(0,T,N+1)
```

```python
#Initalize solutions array
X=np.zeros(N+1)
X[0]=X0
```

```python
#Implement Euler-Maruyama scheme
for i in range(1,N+1):
    dW=np.sqrt(dt)*np.random.normal()
    X[i]=X[i-1]+ mu(X[i-1],t[i-1]) * dt + sigma(X[i-1],t[i-1]) * dW
```

```python
#Plot the results
plt.plot(t,X)
plt.xlabel('Time')
plt.ylabel('X(t)')
plt.title('Euler-Maruyama Approximation of SDE')
plt.show()
```
