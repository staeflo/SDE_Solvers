```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the stochastic process aka drift and diffusion functions
def mu(X,t):
    return 0.1 * X 
def sigma(X,t):
    return 0.3 * X

#Initial condition
X0 = 1.0

#Discretize time
T = 1.0
N = 1_000
dt = T/N
t = np.linspace(0,T,N+1)
# Number of samples
M = 100_000 
```

```python
#Initalize solutions matrix
X=np.zeros((M,N+1))
X[:,0]=X0
```

```python
def euler_maruyama(X0, mu, sigma, T, N, M):
    #Euler-Maruyama scheme
    for i in range(1,N+1):
        dW = np.random.normal(0 , np.sqrt(dt) , M)
        X[:, i] = X[:, i-1] + mu(X[:, i-1], t[i-1]) * dt + sigma(X[:, i-1], t[i-1]) * dW
    return t , X

t, X = euler_maruyama(X0, mu, sigma, T, N, M)
```

```python
# Plot sample paths

for i in range(5):
    plt.plot(t,X [i , :])
plt.xlabel('Time')
plt.ylabel('X(t)')
plt.title('Sample Paths of SDE')
plt.show()
```

```python
#Compute Mean and Variance
mean_X = np.mean(X, axis=0)
var_X = np.var(X, axis=0)
plt.plot(t, np.log(mean_X)/t, label='Mean')
#plt.plot(t, var_X, label='Variance')
plt.xlabel('Time')
plt.ylabel('Value')
#plt.yscale('log')
plt.title('Mean and Variance of SDE')
plt.legend()
plt.show()
```
